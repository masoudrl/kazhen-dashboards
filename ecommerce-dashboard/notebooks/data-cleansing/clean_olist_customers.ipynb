```python
import pandas as pd

# مسیر فایل CSV
file_path = '../../data/raw/olist_customers_dataset.csv'

# بارگذاری داده‌ها
df = pd.read_csv(file_path)

# نمایش ۵ سطر اول
df.head()
```

```python
# بررسی نوع داده‌ها و تعداد مقادیر گمشده
df.info()
print("\nمقادیر گمشده در هر ستون:\n")
print(df.isnull().sum())
```

```python
# حذف سطرهایی که customer_id یا zip_code ندارند (در صورت وجود)
df = df[df['customer_id'].notnull()]
df = df[df['customer_zip_code_prefix'].notnull()]

# یکسان‌سازی ایالت‌ها به حروف بزرگ
df['customer_state'] = df['customer_state'].str.upper()

# حذف فضاهای اضافی از نام شهر
df['customer_city'] = df['customer_city'].str.strip()

# بازنشانی ایندکس
df.reset_index(drop=True, inplace=True)

# پیش‌نمایش داده پاک‌سازی شده
df.head()
```

```python
# ذخیره فایل تمیز شده
clean_path = '../../data/processed/olist_customers_cleaned.csv'
df.to_csv(clean_path, index=False)
print("✅ فایل تمیز شده با موفقیت ذخیره شد.")
```
