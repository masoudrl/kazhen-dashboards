```python
import pandas as pd

file_path = '../../data/raw/olist_products_dataset.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
print("\nMissing values:\n", df.isnull().sum())
```

```python
# حذف سطرهایی که product_id ندارند یا دسته‌بندی نامشخص است
df = df[df['product_id'].notnull()]
df = df[df['product_category_name'].notnull()]

# پر کردن null در وزن یا ابعاد با مقدار میانگین (در صورت وجود)
for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mean(), inplace=True)

df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# ذخیره فایل تمیزشده
clean_path = '../../data/processed/olist_products_cleaned.csv'
df.to_csv(clean_path, index=False)
print("✅ فایل تمیز شده با موفقیت ذخیره شد.")
```
